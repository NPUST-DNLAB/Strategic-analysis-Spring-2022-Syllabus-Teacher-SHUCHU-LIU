```python
import pandas as pd
from opencc import OpenCC
from translate import Translator
```

```python
df = pd.read_excel("Cost_of_Living_Index_2022.xlsx")
```

```python
translator= Translator(to_lang="chinese")
cc = OpenCC('s2twp')
```

```python
a = 0
for d in df['國家']:
    translation = translator.translate(d)
    translation = cc.convert(translation)
    df.at[a,'國家'] = translation
    a += 1
```

```python
cost_of_living_index = 0.21
cost_of_living = []
rent_index = 0.23
rent = []
Grocery_Index = 0.11
Grocery = []
food_price_index = 0.26
food_price= []
local_purchasing_power_index = 0.19
local_purchasing_power =[]
```

```python
for u in df['生活成本指數']:
    cost_of_living.append(u)
for u in df['租金指數']:
    rent.append(u)
for u in df['雜貨指數']:
    Grocery.append(u)
for u in df['食物價指數']:
    food_price.append(u)
for u in df['當地購買力指數']:
    local_purchasing_power.append(u)
```

```python
weight = []
for y in range (len(cost_of_living)):
    weight.append(cost_of_living_index * cost_of_living[y] + rent_index * rent[y] + Grocery_Index * Grocery[y] + food_price_index * food_price[y] + local_purchasing_power_index * local_purchasing_power[y])
```
